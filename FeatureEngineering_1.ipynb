```python
df['state'] = df['state'].map({'isolated':0,'released':1,'deceased' : 2})
```

```python
#storing target variable
Y = df.state

#storing independent variable after eliminating target variable
X = df.drop("state",axis=1)
```

```python
def one_hot_encoding_top_x(X, i, NoOfTopXLabels):
    
    #checking if the column is categorical or not
    if (max(X[i].value_counts().values)>1):
        
            # function to create the dummy variables for the most frequent labels
            # we can vary the number of most frequent labels that we encode
            top_x_labels = [y for y in X[i].value_counts().sort_values(ascending=False).head(NoOfTopXLabels).index]
            print(top_x_labels)
            
            for label in top_x_labels:
                X[i+'_'+label] = np.where(X[i]==label, 1, 0)
    else:
        print("This is not a categorical variable!")
```

```python
one_hot_encoding_top_x(X,'region',4) #passing dataframe, column name and number of top frequent labels required
```
